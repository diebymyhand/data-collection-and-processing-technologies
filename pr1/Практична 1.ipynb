```python
import pandas as pd
import os
```

```python
file_path = 'olist_orders_dataset.csv'
file_size_mbytes = os.path.getsize(file_path) / 1024**2

print(f"File Size: {file_size_mbytes:2f} mbytes")
```

```python
df = pd.read_csv(file_path)
df.head(5)
```

```python
df.info()
```

```python
df_new = df.copy()

columns = df_new.columns.tolist()
columns_date = columns[3:]

for col in columns_date:
    df_new[col] = pd.to_datetime(df_new[col]).dt.as_unit('ns')
```

```python
df_new.info()
```

```python
df_memory = df.memory_usage(deep=True).sum() / 1024**2
df_new_memory = df_new.memory_usage(deep=True).sum() / 1024**2

difference = df_new_memory / df_memory * 100

print(f'Without datetime64[ns]: {df_memory:.2f} mbytes')
print(f'With datetime64[ns]: {df_new_memory:.2f} mbytes')
print(f'\nDifference in memory usage {difference:.2f}%')
```
