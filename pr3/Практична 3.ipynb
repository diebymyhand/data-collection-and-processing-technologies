```python
import numpy as np
from numba import njit
import pandas as pd
import time
from functools import wraps
```

```python
def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        res = func(*args, **kwargs)
        end = time.time()

        total_time = end - start

        print(f'Function {func.__name__!r} took {total_time:.4f} seconds')
        return res
    return wrapper
```

```python
@timer
def numpy_sort(array):
    res = np.sort(array)
    return res
```

```python
@timer
@njit
def numba_sort(array):
    res = array.copy()
    res.sort() 
    return res
```
