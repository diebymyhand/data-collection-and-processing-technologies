```python
import pandas as pd
import numpy as np

np.random.seed(42)
```

```python
def create_dataset(n_rows=1000):
    discounts = [0, 0.05, 0.1, 0.25, 0.50]
    
    data = {
        'product_id': np.arange(1, n_rows+1),
        'quantity': np.random.randint(1, 100, size=n_rows),
        'price': np.round(np.random.uniform(10.0, 500.0, size=n_rows), 2),
        'discount': np.random.choice(discounts, size=n_rows)
    }
    
    df = pd.DataFrame(data=data)
    df['total'] = np.round(df['quantity'] * df['price'] * (1 - df['discount']), 2)
    
    error_indexes = np.random.choice(df.index, size=int(n_rows * 0.1), replace=False)
    df.loc[error_indexes, 'total'] = df.loc[error_indexes, 'total'] * np.random.uniform(1.1, 1.5, size=len(error_indexes))
    
    print(f'Generated {n_rows} rows. Errors: {len(error_indexes)}')
    
    return df
```

```python
df = create_dataset()
```

```python
df.head(10)
```

```python
df['correct_total'] = np.round(df['quantity'] * df['price'] * (1 - df['discount']), 2)

mismatches = df[np.abs(df['total'] - df['correct_total']) > 0.01]

mae = np.abs(df['total'] - df['correct_total']).mean()

print(f'Знайдено невідповідностей: {len(mismatches)}')
print(f'Середня абсолютна різниця (MAE): {mae:.2f}')
```
